```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url = 'insurance.csv'
data = pd.read_csv(url)
data.hist()
plt.show
sns.pairplot(data)

correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'age', y = 'charges', figsize = (9,9), color = 'black')

model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['age']),y= data['charges'])
print(model.intercept_)
print(model.coef_)
# Phuong trinh quy hoi y= coef*x + intercept 
# Do chinh xac
model.score(X = pd.DataFrame(data['age']), y = data['charges'])
y_pred = model.predict(X = pd.DataFrame(data['age']))
plt.plot(data['age'],y_pred,color='blue')

model = linear_model.LinearRegression()
X = pd.DataFrame([data['age'], data['charges']**2]).T
model.fit(X, y = data['charges'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh', model.score(X, y = data['charges']))

range_values = np.arange(1.5, 5.5, 0.1)
data.plot(kind='scatter', x='age', y='charges', figsize=(9, 9), color='black', xlim=(0, 7))
predictor = pd.DataFrame([range_values, range_values**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range_values, y_pred, color='blue')
plt.show()
```
