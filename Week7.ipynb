```python
import pandas as pd
url = 'mtcars.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
correlation= data.corr()
print(correlation)


```

```python

import seaborn as sns
sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
```

```python
from sklearn.metrics import SCORERS
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd

model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']),y= data['mpg'])
print(model.intercept_)
print(model.coef_)
# Phuong trinh quy hoi y= coef*x + intercept 
# Do chinh xac
model.score(X = pd.DataFrame(data['wt']), y = data['mpg'])
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
y_pred = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')

```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T
model.fit(X, y = data['mpg'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh', model.score(X, y = data['mpg']))

range_values = np.arange(1.5, 5.5, 0.1)
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9, 9), color='black', xlim=(0, 7))
predictor = pd.DataFrame([range_values, range_values**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range_values, y_pred, color='blue')
plt.show()
```

```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url_1 = 'Salary_Data.csv'
data = pd.read_csv(url_1)
data.hist()
plt.show
sns.pairplot(data)

correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')

model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']), y = data['YearsExperience'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['Salary']), y = data['YearsExperience'])
y_pred = model.predict(X = pd.DataFrame(data['Salary']))
plt.plot(data['Salary'], y_pred, color='blue')

model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary']]).T
model.fit(X, y = data['YearsExperience'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh', model.score(X, y = data['YearsExperience']))

range_values = np.arange(1.5, 5.5, 0.1)
data.plot(kind='scatter', x='Salary', y='YearsExperience', figsize=(9, 9), color='black', xlim=(0, 7))
predictor = pd.DataFrame([range_values]).T
y_pred = model.predict(X=predictor)
plt.plot(range_values, y_pred, color='blue')

import matplotlib.pyplot as plt
data.plot(kind='scatter',x='YearsExperience',y= 'Salary', figsize=(9,9), color='black')
x = data['YearsExperience']
y = data['Salary']
plt.plot(x,y,'r')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url_3 = 'Advertising.csv'
data = pd.read_csv(url_3)
data.hist()
plt.show
correlation= data.corr()
print(correlation)
sns.heatmap(correlation, annot= True)
sns.pairplot(data)

#Tim su phu thuoc nhau cua cac loai
data.plot(kind='scatter',x='TV',y= 'sales', figsize=(9,9), color='black')
print('MAX(TV): ',data['TV'].max())
print('MIN(TV): ',data['TV'].min())
print('MAX(sales): ',data['sales'].max())
print('MIN(sales): ',data['sales'].min())

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

model = linear_model.LinearRegression()
data = pd.read_csv("Advertising.csv")
X = pd.DataFrame([data['TV'], data['TV']**2]).T
model.fit(X, y = data['sales'])
print('Diem chan cua mo hinh:', model.intercept_)
print('He so cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X, y = data['sales']))

range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize = (9,9), color = 'black', xlim = (0, 300))
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
plt.show()
```

```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url_2 = 'Startups.csv'
data = pd.read_csv(url_2)
data.hist()
plt.show

correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'Profit', y = 'R&D Spend', figsize = (9,9), color = 'black')
sns.pairplot(data)

model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Profit']), y = data['R&D Spend'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['Profit']), y = data['R&D Spend'])
y_pred = model.predict(X = pd.DataFrame(data['Profit']))
plt.plot(data['Profit'], y_pred, color='blue')

model = linear_model.LinearRegression()
X = pd.DataFrame([data['Profit'], data['Profit']**2]).T
model.fit(X, y = data['R&D Spend'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh', model.score(X, y = data['R&D Spend']))

range_values = np.arange(1.5, 5.5, 0.1)
data.plot(kind='scatter', x='Profit', y='R&D Spend', figsize=(9, 9), color='black', xlim=(0, 7))
predictor = pd.DataFrame([range_values, range_values**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range_values, y_pred, color='blue')
plt.show()
```
