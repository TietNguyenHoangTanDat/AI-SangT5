```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = 'Social_Network_Ads.csv'
data = pd.read_csv(url)
data.hist()
plt.show
sns.pairplot(data)
correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'Purchased', y = 'Age', figsize = (9,9), color = 'black')

model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Purchased']), y = data['Age'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['Purchased']), y = data['Age'])
y_pred = model.predict(X = pd.DataFrame(data['Purchased']))
plt.plot(data['Purchased'], y_pred, color='blue')

model = linear_model.LinearRegression()
X = pd.DataFrame([data['Purchased']]).T
model.fit(X, y = data['Age'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh', model.score(X, y = data['Age']))

range_values = np.arange(1.5, 5.5, 0.1)
data.plot(kind='scatter', x='Purchased', y='Age', figsize=(9, 9), color='black', xlim=(0, 7))
predictor = pd.DataFrame([range_values]).T
y_pred = model.predict(X=predictor)
plt.plot(range_values, y_pred, color='blue')

import matplotlib.pyplot as plt
data.plot(kind='scatter',x='Purchased',y= 'Age', figsize=(9,9), color='black')
x = data['Purchased']
y = data['Age']
plt.plot(x,y,'r')
plt.xlabel('Purchased')
plt.ylabel('Age')
plt.legend()
plt.show()
```
